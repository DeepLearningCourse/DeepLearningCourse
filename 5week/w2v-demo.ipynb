```python
from gensim.models import Word2Vec
from nltk.tokenize import word_tokenize
import re
from tqdm import tqdm
import glob
import os
```

```python
tag_clean = re.compile('<[^>]*>', re.U)
```

```python
! ls
```

```python
sentences = []
for file in glob.glob(os.path.join("tolsloy", '*.fb2'))[2:6]\
            + [glob.glob(os.path.join("tolsloy", '*.fb2'))[-2]]:
    print('\n' + file)
    with open(file, encoding='utf-8') as f:
        for line in tqdm(f):
            stripped = line.strip()
            if stripped[:3] == '<p>' and stripped[-4:] == '</p>':
                stripped = tag_clean.sub('', stripped)
                temp = stripped.split(".")
                for t in temp:
                    sentences.append(word_tokenize(t.lower()))
```

```python
sentences[10000:10010]
```

```python
model = Word2Vec(sentences)
```

```python
model["спасибо"]
```

```python
model.most_similar("спасибо")
```
